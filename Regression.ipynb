1. What is Simple Linear Regression?
     - Simple Linear Regression is a statistical technique used to model the linear relationship between one independent variable (X) and one dependent variable (Y) using a straight line

2. What are the key assumptions of Simple Linear Regression?
      - Linear relationship between X and Y
      
      - No significant outliers

      - Independence of observations

      - Normal distribution of errors

      

3.  What does the coefficient m represent in the equation Y=mX+c?
       - The coefficient m represents the slope, showing how much the dependent variable (Y) changes for a one-unit change in X.

4. What does the intercept c represent in the equation Y=mX+c?
      - The intercept c represents the value of Y when X is zero. It provides a baseline value for the relationship.

5.  How do we calculate the slope m in Simple Linear Regression?
      - The slope is calculated as:

        m=Cov(X,Y) / Var(X)

        - It measures the strength and direction of the relationship.
	​


6.  What is the purpose of the least squares method in Simple Linear Regression?
     - The least squares method minimizes the sum of squared residuals, ensuring the best-fitting regression line.

7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?
     - R² indicates the percentage of variance in Y explained by X.

     - For example, R² = 0.8 means 80% of the variation is explained by the model.

8. What is Multiple Linear Regression?
     - Multiple Linear Regression models the relationship between one dependent variable and two or more independent variables

9.  What is the main difference between Simple and Multiple Linear Regression?
     - Simple LR → one predictor

     - Multiple LR → multiple predictors


     - Multiple LR captures more complex relationships.

10.  What are the key assumptions of Multiple Linear Regression?
     - Linearity

     - Independence

     - Homoscedasticity

     - Normal residuals

     - No multicollinearity

11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?
     - Heteroscedasticity occurs when error variance is not constant.
     
    - It leads to unreliable standard errors and hypothesis tests.

12. How can you improve a Multiple Linear Regression model with high multicollinearity?
     - Remove correlated features

     - Use PCA

     - Apply Ridge or Lasso regression

     - Check VIF scores

13. What are some common techniques for transforming categorical variables for use in regression models?
     - One-Hot Encoding

     - Label Encoding

     - Ordinal Encoding

14. What is the role of interaction terms in Multiple Linear Regression?
     - Interaction terms capture how the effect of one variable depends on another, improving model flexibility.

15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?
     - Simple LR: Y when X = 0

     - Multiple LR: Y when all predictors are zero, which may not be meaningful.

16.  What is the significance of the slope in regression analysis, and how does it affect predictions?
     - The slope shows the direction and magnitude of impact of X on Y and directly affects predictions.

17.  How does the intercept in a regression model provide context for the relationship between variables?
     - It sets the baseline level of the dependent variable, helping interpret the regression line.

18. What are the limitations of using R² as a sole measure of model performance?
    - Doesn’t penalize extra variables

    - Can be misleading

    - Doesn’t indicate causality

19.  How would you interpret a large standard error for a regression coefficient?
     - A large standard error means the coefficient estimate is unstable and less reliable.

20.  How can heteroscedasticity be identified in residual plots, and why is it important to address it?
     - Funnel-shaped residuals

     - Increasing spread
    
     - It is important to fix it to ensure valid inference.

21.  What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?
     - It indicates irrelevant predictors are inflating R² without improving the model.

22. Why is it important to scale variables in Multiple Linear Regression?
     - Improves numerical stability

     - Helps gradient-based optimizers

     - Necessary for regularization techniques

23. What is polynomial regression?
     - Polynomial regression models non-linear relationships by adding polynomial terms of predictors.

24. How does polynomial regression differ from linear regression?
     - Linear: straight-line relationship

     - Polynomial: curved relationship

25. When is polynomial regression used?
    - Used when data shows a non-linear trend but still follows a smooth curve.

26.  What is the general equation for polynomial regression?
     - y=a0​+a1​x+a2​x2+⋯+an​xn

27.  Can polynomial regression be applied to multiple variables?
     - Yes, polynomial terms can be added for multiple predictors.

28. What are the limitations of polynomial regression?
    - Overfitting

    - Poor extrapolation

    - Sensitive to noise

29.  What methods can be used to evaluate model fit when selecting the degree of a polynomial?
     - Cross-validation

     - Adjusted R²

     - AIC / BIC

     - Validation error

30. Why is visualization important in polynomial regression?
     - Visualization helps:

     - Identify curvature

     - Detect overfitting

     - Compare model fits

```python
#31. How is polynomial regression implemented in Python?


from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

X = [[1], [2], [3], [4], [5]]
y = [2, 6, 12, 20, 30]

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

predictions = model.predict(X_poly)
print(predictions)
```
